```python
import cv2
import mediapipe as mp
import numpy as np
from google.colab.patches import cv2_imshow

# Initialize MediaPipe Face Mesh
mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh(static_image_mode=True, max_num_faces=1)

# Load image
image = cv2.imread("000001.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Detect face landmarks
results = face_mesh.process(image_rgb)

if not results.multi_face_landmarks:
    print("No face detected!")
    exit()

# Extract landmark coordinates
h, w, _ = image.shape
landmarks = results.multi_face_landmarks[0].landmark
points = [(int(p.x * w), int(p.y * h)) for p in landmarks]

# Nose landmark indices
nose_indices = [1, 2, 4, 5, 168, 94, 195, 331, 6]
nose_points = np.array([points[i] for i in nose_indices], dtype=np.int32)

# Draw original nose landmarks
image_original = image.copy()
for x, y in nose_points:
    cv2.circle(image_original, (x, y), 2, (0, 255, 0), -1)

# Make a copy of the image for warping
warp = image.copy()

# Define reshaping - "Narrow & Pointed"
# Let's move side points (195, 331) slightly inward to narrow the nose
reshaped_points = nose_points.copy()
reshaped_points[6][0] -= 10  # move point 195 left
reshaped_points[7][0] += 10  # move point 331 right
reshaped_points[2][1] -= 5   # move point 4 (tip) up to make it pointier

# Create Delaunay triangulation for warping
rect = (0, 0, w, h)  # Full image dimensions
subdiv = cv2.Subdiv2D(rect)
for p in reshaped_points:
    x, y = float(p[0]), float(p[1])
    if 0 <= x < w and 0 <= y < h:  # Make sure point is inside image bounds
        subdiv.insert((x, y))


triangles = subdiv.getTriangleList().astype(np.int32)

# Warp each triangle
for tri in triangles:
    pts1 = []
    pts2 = []
    for i in range(0, 6, 2):
        x, y = tri[i], tri[i+1]
        if (x, y) in nose_points.tolist():
            idx = nose_points.tolist().index((x, y))
            pts1.append(nose_points[idx])
            pts2.append(reshaped_points[idx])

    if len(pts1) == 3:
        warp_mat = cv2.getAffineTransform(np.float32(pts1), np.float32(pts2))
        triangle_area = cv2.boundingRect(np.float32([pts2]))
        x, y, w_t, h_t = triangle_area
        roi = image[y:y+h_t, x:x+w_t]
        mask = np.zeros((h_t, w_t, 3), dtype=np.uint8)
        pts2_roi = np.array([[p[0]-x, p[1]-y] for p in pts2], dtype=np.int32)
        cv2.fillConvexPoly(mask, pts2_roi, (1, 1, 1))
        warped = cv2.warpAffine(image, warp_mat, (w, h))
        roi_warped = warped[y:y+h_t, x:x+w_t]
        roi_result = roi * (1 - mask) + roi_warped * mask
        warp[y:y+h_t, x:x+w_t] = roi_result

# Show before and after
cv2_imshow(image_original)  # Original nose with landmarks
cv2_imshow(warp)            # Narrow & Pointed reshaped nose
```
